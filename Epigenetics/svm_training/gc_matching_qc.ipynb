```python
import os
import subprocess
import pandas as pd
```

## Check GC-matched negatives for training examples:

```python
gc_diff_dict_train = {'cluster': [i for i in range(1, 25) for j in range(10)], 'fold': [j for i in range(1, 25) for j in range(10)], 'diff_avg_gc': [], 'avg_diff_gc': []}
for i in range(1, 25):
    cluster = 'Cluster' + str(i)
    print(cluster)
    for j in range(10):
        gc_dict = {'pos':[], 'neg':[]}
        fold = 'fold' + str(j)
        pos = '/mnt/lab_data3/soumyak/adpd/clusters_gkmsvm/'+cluster+'/'+fold+'/train/train.final.pos.fasta'
        neg = '/mnt/lab_data3/soumyak/adpd/clusters_gkmsvm/'+cluster+'/'+fold+'/train/train.final.neg.fasta'
        pos_seqs = [i.strip().upper() for i in open(pos).readlines() if not i.startswith('>')]
        neg_seqs = [i.strip().upper() for i in open(neg).readlines() if not i.startswith('>')]
        pos_gc = sorted([(seq.count('G') + seq.count('C')) / len(seq) for seq in pos_seqs], reverse=True)
        neg_gc = sorted([(seq.count('G') + seq.count('C')) / len(seq) for seq in neg_seqs], reverse=True)
        gc_dict['pos'] = pos_gc
        gc_dict['neg'] = neg_gc
        gc_df = pd.DataFrame.from_dict(gc_dict)
        gc_df['diff'] = abs(gc_df['pos'] - gc_df['neg'])
        mean_pos_gc = gc_df['pos'].mean()
        mean_neg_gc = gc_df['neg'].mean()
        diff_avg_gc = abs(mean_pos_gc - mean_neg_gc)
        avg_diff_gc = gc_df['diff'].mean()
        gc_diff_dict_train['diff_avg_gc'].append(diff_avg_gc)
        gc_diff_dict_train['avg_diff_gc'].append(avg_diff_gc)
        #display(gc_df.head())
        #display(gc_df.tail())
```

```python
gc_diff_df_train = pd.DataFrame.from_dict(gc_diff_dict_train)
display(gc_diff_df_train)
display(gc_diff_df_train.shape)
display(gc_diff_df_train.mean())
```
